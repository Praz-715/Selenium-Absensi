```python
import time
import random
import os
import requests
import aiohttp
import asyncio
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from PIL import Image

# Path to your ChromeDriver
service = Service("path_to_your_chromedriver")  # Adjust your ChromeDriver path

# Asynchronous function to upload the picture
async def upload_pic(image_name, cookie_string):
    url = "https://www.vankasystem.net/absensi/ajax/selfi"
    file_path = os.path.abspath(os.path.join("images", image_name))
    
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")
    
    headers = {
        "Cookie": cookie_string,  # Sending the cookies for authentication
    }
    
    # Opening file asynchronously and uploading it using aiohttp
    async with aiohttp.ClientSession() as session:
        with open(file_path, "rb") as f:
            # Preparing the form-data for file upload
            form_data = aiohttp.FormData()
            form_data.add_field('webcam', f, filename=image_name)
            
            async with session.post(url, headers=headers, data=form_data) as response:
                if response.status == 200:
                    return await response.text()  # Return the response text (image filename)
                else:
                    raise Exception(f"Upload failed with status {response.status}: {await response.text()}")

# Function to log in and handle absence submission
async def login_and_absence():
    options = Options()
    options.add_argument("--start-maximized")
    options.add_argument("--disable-notifications")
    
    # Initialize WebDriver
    driver = webdriver.Chrome(service=service, options=options)
    driver.get("https://www.vankasystem.net/absensi/login")

    # Wait for page to load
    time.sleep(3)

    # Perform login
    username = driver.find_element(By.NAME, "username")
    password = driver.find_element(By.NAME, "password")
    username.send_keys("ali")  # Change username
    password.send_keys("superman")  # Change password
    password.send_keys(Keys.RETURN)

    # Wait for the page to load after login
    time.sleep(5)

    # Get cookies after login
    cookies = driver.get_cookies()
    cookie_string = "; ".join([f"{cookie['name']}={cookie['value']}" for cookie in cookies])
    print(f"Cookies: {cookie_string}")

    # Random picture selection
    random_picture = get_random_picture()

    # Upload picture asynchronously
    try:
        upload_response = await upload_pic(random_picture, cookie_string)
        print("Upload response:", upload_response)
    except Exception as e:
        print(f"Error during picture upload: {e}")
        return

    # Perform the attendance submission
    attendance_result = await submit_attendance(upload_response, cookie_string)
    print("Attendance result:", attendance_result)

    # Close the browser after operations
    time.sleep(5)
    driver.quit()

# Function to get a random picture for upload
def get_random_picture():
    pictures = [
        "1724131171.png", "1708052252.png", "1708686358.png", 
        "1709258931.png", "1709518168.png", "1709550829.png",
        "1709867274.png", "1710729449.png", "1710761084.png", 
        "1710848697.png", "1710987038.png"
    ]
    return random.choice(pictures)

# Asynchronous function to submit attendance
async def submit_attendance(upload_response, cookie_string):
    data = {
        "maps_absen": "-6.1855254,106.8023633",  # Example geolocation
        "base64image": upload_response  # Assuming upload returns base64 image or image filename
    }

    headers = {
        'Accept': 'application/json, text/javascript, */*; q=0.01',
        'Cookie': cookie_string,
        'Content-Type': 'application/x-www-form-urlencoded; charset=UTF-8'
    }

    try:
        async with aiohttp.ClientSession() as session:
            async with session.post("https://www.vankasystem.net/absensi/ajax/absenajaxnew", data=data, headers=headers) as response:
                if response.status == 200:
                    return await response.json()  # Return response as JSON
                else:
                    raise Exception(f"Attendance submission failed with status {response.status}: {await response.text()}")
    except Exception as e:
        print(f"Error during attendance submission: {e}")
        return None

# Main execution point
if __name__ == "__main__":
    asyncio.run(login_and_absence())
```
